#### **Programa para convertir de kilómetros a millas**

#### **1 km = 0.621371 millas**

```python
# solicitar los kilometros
k = float(input("Dame el número de kilómetros: "))
# 1 km = 0.621371 millas
factor = 0.621371
# calcular las millas
millas = k * factor
print("El valor en millas es:  %.3f " %(millas) )
```

### **Uso de condicionales IF**
##### **Pedimos el valor de kilómetros que se corrieron, los convertimos a millas**
##### **Si el valor de las millas es = 10 entonces imprimir "Meta no cumplida"**
##### **de otra forma impimir "Meta cumplida"**

```python
# solicitar los kilometros
k = float(input("¿Cuántos kilometros corriste ? "))
# 1 km = 0.621371 millas
factor = 0.621371
# calcular las millas
millas = k * factor

# le faltan los dos puntos en el else
if millas = 10:
    print("Meta no cumplida")
else
    print("Meta cunplida")

# imprimir una salida como esta: 1 Kilometro(s) equivale a .621 millas
# print("%.1f Kilometro(s) equivale a: %.2f millas " %(k, millas) )
```
